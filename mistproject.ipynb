```python
sent = re.compile(r'[!?.\n]')
sent = sent.split(text)
sent
```

```python
clean_sent = [] # почистили от лишних пробелов / строк
for i in sent:
  if len(i.strip()) > 0:
    clean_sent.append(i.strip())
clean_sent
```

```python
keys = re.findall(r'\w* *еха. \D\w*', text) # пока в исходном тексте, чтобы быстро получить ключи
keys
```

```python
keys = list(set(keys))
keys # удалили повторы, т.к. ключи должны быть уникальными
```

```python
mistakes = {}

for k in keys:
  l = []
  for i in clean_sent:
    if k in i: # т.е. если нужные словосочетания есть в строке
      l.append(i)
  mistakes[k] = l
  mistakes
```

```python
for i in list(mistakes.keys()): # просто напечатаю по-другому, чтобы было лучше видно ключи - значения
  print(i)
  print(*mistakes[i], '\n', sep='\n')
```
